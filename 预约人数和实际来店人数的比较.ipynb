```python
df_air_reserve['visit_time'] = df_air_reserve.visit_datetime
df_air_reserve['reserve_time'] = df_air_reserve.reserve_datetime

def get_date(datetime):
    return datetime.split(' ')[0]

def get_time(datetime):
    return datetime.split(' ')[1]

df_air_reserve.visit_time = df_air_reserve.visit_time.apply(lambda x: get_time(x))
df_air_reserve.reserve_time = df_air_reserve.reserve_time.apply(lambda y: get_time(y))

df_air_reserve.visit_datetime = df_air_reserve.visit_datetime.apply(lambda x: get_date(x))
df_air_reserve.reserve_datetime = df_air_reserve.reserve_datetime.apply(lambda y: get_date(y))
```
